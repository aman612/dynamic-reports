```python
"""population of Canada"""
data = px.data.gapminder()
data_canada = data[data.country == 'Canada']
fig = px.bar(data_canada, x='year', y='pop',
             hover_data=['lifeExp', 'gdpPercap'], color='lifeExp',
             labels={'pop':'population of Canada'}, height=400)
fig.show()
```

```python
"""population of Canada"""
data = px.data.gapminder()
data_canada = data[data.country == 'Canada']
fig = px.bar(data_canada, x='year', y='pop',
             hover_data=['lifeExp', 'gdpPercap'], color='lifeExp',
             labels={'pop':'population of Canada'}, height=400)
fig.show()

pio.write_image(fig, 'output_data/image/canada_population.png')
```

```python
#creating a new excel file named 'report.xlsx' to avoid data loss incase
writer = pd.ExcelWriter('output_data/files/report.xlsx', engine='xlsxwriter')

#save the updated data to an excel file with sheet name 'Sales Data'
data_canada.to_excel(writer, index=False, sheet_name='Canada Population Data')



# workbook = writer.book
# worksheet = writer.sheets["Sheet1"]

#creating a worksheet object named 'worksheet' within the Excel file that was created earlier using the 'ExcelWriter' function.
worksheet = writer.sheets['Canada Population Data']

# add the chart to the newly created Excel file
worksheet.insert_image('K1', 'output_data/image/canada_population.png')

# #Finally save the file
writer._save()
```
